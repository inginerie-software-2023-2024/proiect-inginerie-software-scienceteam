```python
import cv2
import numpy as np
import albumentations as A

def augment_image(img):
    transform = A.Compose([
        A.HorizontalFlip(p=0.7),
        A.RandomContrast(p=0.5),
        A.OneOf([
            A.MotionBlur(p=0.3),
            A.MedianBlur(blur_limit=3, p=0.3),
            A.Blur(blur_limit=3, p=0.3),
            A.AdvancedBlur(p=0.3),
            A.GaussianBlur(p=0.3),
            A.ZoomBlur(p=0.3),
        ], p=0.2),
        A.ShiftScaleRotate(p=0.3),
        A.ChannelShuffle(p=0.5),
        A.RandomRain(p=0.2),
        A.RandomBrightness(p=0.5),
        A.RandomRotate90(p=0.4),
        A.RandomRotate90(p=0.3),
        A.VerticalFlip(p=0.6),
        A.HueSaturationValue(p=0.3),
        A.RGBShift(p=0.4),
        A.GaussNoise(p=0.2),
        A.Downscale(p=0.1),
        A.InvertImg(p=0.2),
        A.ISONoise(p=0.3),
        A.PixelDropout(p=0.2),
    ])
    img = transform(image=img)['image']
    
    return img
```

```python
from PIL import Image

base_directory_path = '/home/mircea/IS/50classes_augmented1000/ip102_v1.1/images/'
train_txt_directory_path = '/home/mircea/IS/50classes_augmented10000/ip102_v1.1/train.txt'
temporal_images_directory = 'temporal_images/'
temporal_adnotations_directory = 'temporal_adnotations/'
starting_index = 146099
f = open('temporal_adnotations/temporal_train.txt','w')
from tqdm import tqdm
for cls in tqdm(images_per_minority_class.keys()):
    no_augmented_images_for_class = 0
    while continue_loop(cls):
        for img_file in images_per_minority_class[cls]:
            
            img = Image.open(base_directory_path+img_file)
            img = img.resize((224,224))
            img = img.convert('RGB')
            img = np.array(img)
            augmented_image = augment_image(img)
            augmented_image = Image.fromarray(augmented_image)
            augmented_image = augmented_image.convert('RGB')
            
            augmented_image.save(temporal_images_directory+str(starting_index)+'.jpg')
            
            f.write(str(starting_index)+'.jpg'+' '+str(cls)+'\n')
            
            starting_index += 1
            examples_per_minority_class[cls] += 1
            no_augmented_images_for_class += 1
            if not continue_loop(cls):
                break
    print(f'Class {cls} has {no_augmented_images_for_class} augmented images')

f.close()
```

Calculate class weights

```python
def class_weights(labels):
    c = Counter(labels)
    mc = c.most_common()
    weights = []
    total_samples = len(labels)

    # option 1:
    # nr of samples divided by the number of samples for each class
    for i in range(len(mc)):
        weights.append(total_samples/ mc[i][1])
        
    # option 2: 1-no_samples_per_class/total_samples
        
        
    return weights
```
